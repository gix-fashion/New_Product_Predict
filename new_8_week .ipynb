```python
import pandas as pd
import matplotlib.pyplot as plt
#data_xls = pd.read_excel('shirts.xlsx')
#data_xls.to_csv('shirts.csv',encoding='utf-8')
df = pd.read_csv('shirts.csv',encoding='utf-8',low_memory=False)
df = df.drop(df.columns[0],axis = 1)
```

```python
#增加地区特征
df['地区'] = df.apply(lambda x: x[2].strip('分公司'),axis =1)
# 删除分公司,销售金额，库存数量信息
df = df. drop(df.columns[[2,3,8,9,11,12,14,15,16]],axis = 1)
df = df.dropna()
```

```python
#创建第n周销量
df['第一次销售年份'] = 0
df['第一次销售周次'] = 0
df['第1周销量'] = 0
df['第2周销量'] = 0
df['第3周销量'] = 0
df['第4周销量'] = 0
df['第5周销量'] = 0
df['第6周销量'] = 0
df['第7周销量'] = 0
df['第8周销量'] = 0
df['是否总结行'] = df.apply(lambda x: x[6]==x[7],axis = 1)
df['第一次销售年份'] = df.apply(lambda x : x[19]*x[0],axis = 1)
df['第一次销售周次'] = df.apply(lambda x : x[19]*x[1],axis = 1)
df['第1周销量'] =  df.apply(lambda x : x[19]*x[6],axis = 1)
```

```python
# 把同一种商品的销售量按时间放在相邻行
df = df.sort_values(by = ['小类','波段','商品年份','季节','地区','累计销售数量'])
df = df.reset_index(drop= True)
df.head(10)
```

```python
# 收集总结行索引值
sum_indexes = df[df.是否总结行].index.tolist()
for index in sum_indexes:
    df.loc[index,'第2周销量'] = df.loc[index+1,'本周销售数量']
    df.loc[index,'第3周销量'] = df.loc[index+2,'本周销售数量']
    df.loc[index,'第4周销量'] = df.loc[index+3,'本周销售数量']
    df.loc[index,'第5周销量'] = df.loc[index+4,'本周销售数量']
    df.loc[index,'第6周销量'] = df.loc[index+5,'本周销售数量']
    df.loc[index,'第7周销量'] = df.loc[index+6,'本周销售数量']
    df.loc[index,'第8周销量'] = df.loc[index+7,'本周销售数量']
df =df[df.是否总结行]
df['8周总销量'] = df.apply(lambda x : x[11]+x[12]+x[13]+x[14]+x[15]+x[16]+x[17]+x[18],axis =1)
df.head(10)
```

```python
df = df.reset_index(drop= True)
# 删除不必要列
df = df. drop(df.columns[[0,1,6,7,15,16,17,18,19]],axis = 1)
df.head(10)
```
